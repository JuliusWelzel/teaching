```python
import pandas as pd
import numpy as np
from scipy import stats as st
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
# import data
# first import the data and visualize content


raw = pd.read_csv('Feedback_funktionelle_neuroanatomie.csv')
print(raw.columns)

ax = sns.histplot(
    raw.iloc[:,1:6],multiple="dodge", shrink=.8,discrete=True
)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
pre = raw["Ich verstehe die Paper vor der Seminarsitzung gut"]
post = raw["Ich verstehe die Paper nach der Seminarsitzung gut"]

m_pre = np.mean(pre)
sd_pre = round(np.std(pre),2)

m_post = np.mean(post)
sd_post = round(np.std(post),2)

df = len(pre)
t_, p_ = st.ttest_rel(pre,post)

# apa format ttest results
print("The results from pre-seminar (M = {}, SD = {}) and post-seminar (M = {}, SD = {}) indicate that seminar attendance did not resulted in an improvement in paper understanding, t({}) = {}, p = {}.".format(m_pre,sd_pre,m_post,sd_post,df,round(t_,3),round(p_,3)))
```
